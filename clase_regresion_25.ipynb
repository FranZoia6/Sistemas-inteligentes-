```python
# Estandarizamos:
mms = MinMaxScaler()
mms.fit(X_train)
training_dataset_x = mms.transform(x_train)
test_dataset_x = mms.transform(x_test)
valid_dataset_x = mms.transform(x_valid)
```

```python
# Estandarizamos:
std = StandardScaler()
std.fit(X_train)
training_dataset_x = std.transform(X_train)
test_dataset_x =std.transform(x_test)
valid_dataset_x = std.transform(x_valid)
```

```python
plt.hist(x=training_dataset_x, rwidth=0.85)
plt.title('- Histograma de valores estandarizados - ')
plt.xlabel('values stz')
plt.ylabel('Freq')
plt.show() #dibujamos el histograma
```

El modelo que utilizaremos es un modelo de regresin lineal regularizado, conocido como Ridge regression. Lo que s tenemos que saber ahora es que este modelo es capaz de resolver un problema de regresin lineal, y que tiene un hiperparmetro que usualmente se llama lambda pero que en Sklearn se denota con alpha, que controla el nivel de regularizacin. La regularizacin es una estrategia matemtica que se utiliza durante el entrenamiento para evitar el overfitting. Idealmente, su utilizacin nos permite mejorar la capacidad de generalizacin del modelo. Sin embargo, a veces puede ocurrir que la utilicemos y que los resultados no mejoren.

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
model_ridge = Ridge(alpha=0.01,solver='sag') # definimos el parmetro alpha=0.01 y la funcin solver
```

```python
# Entrenamos el modelo:
model_ridge.fit(training_dataset_x,Y_train)
```

```python
# Predict:
y_pred = model_ridge.predict(valid_dataset_x)
```

```python
# Calculamos el error cuadrtico medio y el estadstico R^2
print('Error cuadrtico medio: %.2f' % mean_squared_error(y_valid,y_pred))
print('RMSE: %.2f' % (mean_squared_error(y_valid,y_pred)**0.5))
print('MAE: %.2f' % mean_absolute_error(y_valid,y_pred))
print('R_2: %.2f' % r2_score(y_valid,y_pred))
```

```python
# Entrenamos el modelo:

model2_lasso = Lasso(alpha=0.1) # definimos el parmetro alpha=0.1 y la funcin solver
model2_lasso.fit(training_dataset_x,Y_train)
```

```python
# predict:
pred_valid = model2_lasso.predict(test_dataset_x)
```

```python
# Calculamos las mtricas:
print('Error cuadrtico medio: %.2f' % mean_squared_error(y_test, pred_valid))
print('RMSE: %.2f' % (mean_squared_error(y_test, pred_valid)**0.5))
print('MAE: %.2f' % mean_absolute_error(y_test, pred_valid))
print('R_2: %.2f' % r2_score(y_test, pred_valid))
```

## **Random Forest Regressor**



Random Forest es un modelo de machine learning basado en ensambles que nos permite hacer tanto regresin como clasificacin. Los modelos basados en ensambles integran las respuestas de varios submodelos para producir una respuesta final. En el caso de RF, se cuenta con varios rboles de decisin entrenados a partir de particiones aleatorias de los datos de entrenamientos y subconjuntos de caractersticas.

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Instanciamos el modelo
rfr = RandomForestRegressor(n_estimators=30)
```

```python
# Entrenamos:
rfr.fit(training_dataset_x,Y_train)
```

```python
# Prediccin
y_pred = rfr.predict(test_dataset_x)
```

```python
# Calculamos las mtricas de regresin:
print('Error cuadrtico medio: %.2f' % mean_squared_error(y_test, y_pred))
print('RMSE: %.2f' % (mean_squared_error(y_test, y_pred)**0.5))
print('MAE: %.2f' % mean_absolute_error(y_test, y_pred))
print('R_2: %.2f' % r2_score(y_test, y_pred))
```
